```python
def clean_age(dob, age):
        if dob > datetime.now() and age < 100 :
            new_dob = dob.replace(year=datetime.now().year - age)
        if age >= 100 and dob.year > datetime.now().year - 100:
            new_dob = dob.replace(year=datetime.now().year - age)
        else:
            return dob
        return new_dob
```

```python
def clean_cols(df: pd.DataFrame) -> pd.DataFrame:
    df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns] 
    return df
```

```python
def hash_password(pw):
    if pd.isna(pw):
        return None
    return hashlib.sha256(str(pw).encode('utf-8')).hexdigest()
```

```python
def clean_salary(salary: str) -> str:
    if salary is None:
        return None
    else:
        return re.sub(r'[^\d.]', '', salary) 
```
